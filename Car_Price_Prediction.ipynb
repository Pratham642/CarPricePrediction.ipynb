#Preprocessing

```python
df['doornumber'].unique()
```

```python
df['doornumber']=df['doornumber'].map({'two':2,'four':4})
```

```python
X=df.drop('car_ID',axis=1)
y=df['car_ID']
```

```python
X['fuel']=X.fueltype.map({'gas':0,'disel':1})
```

```python
X['engine']=X.enginelocation.map({'front':0,'rear':1})
```

```python
X=df.drop('fueltype',axis=1)
```

```python
X=df.drop('enginelocation',axis=1)
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
X.iloc[:,0]=le.fit_transform(X.iloc[:,0])
```

```python
categorical_cols = ['CarName', 'fueltype', 'aspiration', 'doornumber', 'carbody',
                   'drivewheel', 'enginetype', 'cylindernumber', 'fuelsystem']

# Apply label encoding to each categorical column
```

```python
for col in categorical_cols:
    X[col] = le.fit_transform(X[col])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

Now,there is regresion problem and target is numeric so we will use 1.Linear regression: as the data is small 2.descision tree: as the data is small and majority feature are indepnedent& there are non-linear relationship between feature and target 3.random forest:-to handle the robust method

why not to use 1.logistic regression: the target is numeric and we use logistic when target is binary 2.knn:-as it is lazy algorithm and requires more time for execution

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
dsc=DecisionTreeClassifier()
```

```python

dsc.fit(X_train, y_train)
```

```python
y_pred_dsc = dsc.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
def evaluate_model(y_test, y_pred, DecisionTreeRegressor):
    print(f"--- {DecisionTreeRegressor} Evaluation ---")
    print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
    print(f"R-squared Score: {r2_score(y_test, y_pred)}")
```

```python
evaluate_model(y_test, y_pred_dsc, "DecisionTreeRegressor")
```
