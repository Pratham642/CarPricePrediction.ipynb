```python

#Step 2: Define the list educational instiute paragraph (context)
paragraph="""
Hugging Face is a leading platform and company in the field of artificial intelligence, specializing in natural language processing (NLP) and
machine learning. Known for its open-source libraries like Transformers, Hugging Face provides pre-trained models, tools, and frameworks for tasks such as text generation,
sentiment analysis, translation, and more. It fosters collaboration and innovation by hosting a model hub where researchers and developers can share and deploy AI models.
With a focus on accessibility and user-friendliness, Hugging Face is a cornerstone of the AI community, empowering individuals and organizations to build cutting-edge applications."""

#step 3: Define the list of questions to ask
questions=[
    "What is Hugging face?",
    "What type of models are there in hugging face?",
    "What does hugging face focuses on?",
    "What is the purpose of Hugging Face's model?",
    "How does Hugging Face contribute to the AI community?"
]
#step 4": ask questions and get answers from the paragraph
for question in questions:
  result=qa_pipeline(question=question,context=paragraph)
  print(f"Q:{question}")
  print(f"Q:{result['answer']}\n")






```

```python
from transformers import pipeline
import PyPDF2
#Step 1: Initialize the Question Anwering pipeline with an explicit model
qa_pipeline=pipeline ("question-answering",model="distilbert-base-cased-distilled-squad",tokenizer="distilbert-base-cased")
```

```python
#step 2: Function to extract text from a PDF  file
def extract_text_from_pdf(pdf_path):
  pdf_text=""
  with open(pdf_path,"rb")as file:
    reader=PyPDF2.PdfReader(file)
    for page in reader.pages:
      pdf_text=pdf_text+page.extract_text()+"\n" #Append text from
      return pdf_text
```

```python
#Step 3:Define the pdf file path
pdf_path="/content/drive/MyDrive/Dataset/Mumbai University.pdf"
context=extract_text_from_pdf(pdf_path)
```

```python
context
```

```python

#Step 4:Define questions to ask based on the PDF content
questions=[
"When was Mumbai University established, and what is its historical significance?",
"What are the main campuses of Mumbai University, and what is unique about each?",
"What types of programs and courses does Mumbai University offer?",
"How does Mumbai University support research and innovation?",
"Name some popular programs introduced by Mumbai University that cater to industry needs."
]
```

```python

#STep 5: Ask questions and get answers from the PDF content
print("Extracted Questions and Answers from PDF:\n")
for question in questions:
  result=qa_pipeline(question=question, context=context)
  print(f"Q: {question}")
  print(f"A: {result['answer']}\n")
```

#Sentiment Analysis

```python
from transformers import pipeline

#Load the sentiment analysis pipeline with a small, efficient model
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
sentiment_pipeline
```

```python
#Sample texts from sentiment analysis
texts = [
    "I love this product! It's amazing",
    "The experience was terrible, I hated it",
    "It's okay, not bad but not good either"
]
```

```python
#Perform sentiment analysis
results=sentiment_pipeline(texts)
```

```python
results
```

```python
#Print Results(ca can be trated as an option)
for text, result in zip(texts, results):
  print(f"Text: {text}\nSentiment: {result['label']}(Confidence: {result['score']:.2f})\n")
```
