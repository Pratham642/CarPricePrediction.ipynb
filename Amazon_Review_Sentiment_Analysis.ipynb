```python
corpus["polarity_scores"]=corpus["review"].apply(lambda review:se.polarity_scores(review))
```

```python
corpus
```

```python
corpus["compound"]=corpus['polarity_scores'].apply(lambda x:x['compound'])
```

```python
corpus
```

```python
age=21
print('You are eligible to vote') if(age>=18) else print('You are not eligible to vote')
```

```python
corpus['sentiment']=corpus['compound'].apply(lambda x:'pos' if x>0 else 'neg')
```

```python
 corpus
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
y_true=corpus['label']
y_pred=corpus['sentiment']
```

```python
confusion_matrix(y_true,y_pred)
```

```python
print(classification_report(y_pred,y_true))
```
